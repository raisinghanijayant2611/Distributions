#### Bernoulli and Binomial Distribution

These are the two distributions which are generally used to define the pdf of a <u>Binary Random Variable </u>. A binary Random Variable is a variable which is associated with an event having only two outcomes. For ex. tossing a coin is an event having only 2 outcomes. Thus, if I describe a Random Variable X = 1 whenever outcome is "heads", the probability of X=1 in given N number of trials can be formulated using such distributions 

#### Difference between Bernoulli and Binomial 

Bernoulli Random Variable corresponds to just one event which has two possible outcomes. For ex. an event corresponding to tossing a coin just one time. Suppose the probability of "Heads" as an output is <b><l>u</b></l>. Now suppose I define a Random Variable X = 1 corresponding to the event of occuring heads, such a RV representing the result coming out to be a Head is a bernoulli RV.  


On the other hand, the collection of Bernoulli RV constitutes Bionomial RV. For ex. an event of tossing a coin only once corresponds to Bernoulli RV but an event to get 3 "Heads" in 5 trails can be represented by Binomial RV. 

Thus, every Bernoulli RV can be represented as a Binomial RV but opposite is not true always 

#### Probability Mass Functions of Bernoulli and Binomial 

The above event can be expressed as an equation as follows: 
- The probability of occuring "Heads", or RV X =1 is <b>u</b> 
- The probability of occuring "Tails" is <b>(1-u)</b>


Thus, the probability of x given <b>u</b> or the probability distribution over x is given by : 

\begin{align*}
{u^{x}}{(1-u)^{(1-x)}}
\end{align*}




I would like to differente between "Probability Distribution", "Probability Distribution Function" and "Probability Mass Function" here. It might look easy to majority of audience but I have observed some people getting confused upon these terms being used together. The probability distribution is the collection of all the probability outcomes correponding to every value of Random Variable. The probability distribution function is just  a function applied on that distribution which gives you the value or probability of the Random Variable X corresponding to its particular value (X = x). The probability mass function is similar to pdf, only difference being that pmf is used for discrete distributions whereas pdf is used for continuous distributions. The reason for this is being well explained in the [link](https://math.stackexchange.com/questions/23293/probability-density-function-vs-probability-mass-function)


Thus, the above equation corresponds to Bernoulli Event which is a discrete event having only two outcomes "Heads" or "Tails". We can also write it as a function of x, generally called "Probability Mass Function" of x as follows : 


\begin{align*}
f(x,u) = {u^{x}}{(1-u)^{(1-x)}}
\end{align*}


On the other hand, if we extend this same event, i.e. suppose if we toss the coin N times, the probability of getting heads m times is given by : 

\begin{align*}
P(X=m) = {N \choose m}{u^{m}}{(1-u)^{(N-m)}}
\end{align*}

The above equation is of <b>Binomial distribution</b> which is modelled upon bernoulli trails. This distribution is also a discrete distribution. Though the values can vary upon a certain range instead of just 2 fixed outcomes, however if you plot them in a graph, there would be individual points not touching each other. #### Expected value and variance of Binomial distribution

Following is the derivation of mean of Binomial distribution:


The pdf can be written as : 

\begin{align*}
f(N,x,u) = {N \choose x}{u^{x}}{(1-u)^{(N-x)}}
\end{align*}

The mean of a distribution is defined in terms of expected value, which is the weighted average of all the possible values of the RV, and the weights are given by their respective probabilities. Thus, the equation then becomes : 

\begin{align*}
\sum_{x=1}^n xp(x)
\end{align*}

Thus, the expected value of Binomial Distribution is given by : 

\begin{align*}
\sum_{x=1}^n x{n \choose x}{u^{x}}{(1-u)^{(n-x)}}
\end{align*}


\begin{align*}
= \sum_{x=1}^n x{\frac{(n)!}{(x)!(n-x)!}}{u^{x}}{(1-u)^{(n-x)}}
\end{align*}



\begin{align*}
= \sum_{x=1}^n x{\frac{n(n-1)!}{(x)!(n-x)!}}{u*u^{x-1}}{(1-u)^{(n-x)}}
\end{align*}


\begin{align*}
= nu\sum_{x=1}^n x{\frac{(n-1)!}{x*(x-1)!(n-x)!}}{u^{x-1}}{(1-u)^{(n-x)}}
\end{align*}


\begin{align*}
= nu\sum_{x=1}^n {\frac{(n-1)!}{(x-1)!((n-1)-(x-1))!}}{u^{x-1}}{(1-u)^{((n-1)-(x-1))}}
\end{align*}


Substituting (n-1) as a , x-1 as y. The above equation becomes : 

\begin{align*}
= nu\sum_{y=0}^a {\frac{(a)!}{(y)!(a-y)!}}{u^{y}}{(1-u)^{((a-y))}}
\end{align*}

\begin{align*}
= nu((u) + (1-u))^a
\end{align*}


\begin{align*}
= nu
\end{align*}



Similarly, the variance can be derived in a similar way. We know that it is defined as the expected value of difference between the RV and its mean 

\begin{align*}
Var(X) = E[(X-u)^2]
\end{align*}


\begin{align*}
= \sum_{x=1}^n(x-u)^2p(x)
\end{align*}


Since expectation of an expected value is the same expected value. The variance can also be written as : 

\begin{align*}
= E[X^2] - [E[X]]^2  
\end{align*}

We first derive the Expectation of square of X minus X and then come to the equation above. 
Thus, 


\begin{align*}
E[X^2 - X] = \sum_{x=1}^n(x^2-x)p(x)
\end{align*}


\begin{align*}
= \sum_{x=1}^n(x^2-x){\frac{(n)!}{(x)!(n-x)!}}{u^{x}}{(1-u)^{(n-x)}}
\end{align*}

\begin{align*}
= \sum_{x=1}^n(x)(x-1){\frac{(n)!}{(x)(x-1)(x-2)!(n-x)!}}{u^{x}}{(1-u)^{(n-x)}}
\end{align*}


\begin{align*}
= \sum_{x=1}^n{\frac{(n)(n-1)(n-2)!}{(x-2)!(n-x)!}}{u^{x}}{(1-u)^{(n-x)}}
\end{align*}

\begin{align*}
= (n)(n-1)\sum_{x=1}^n{\frac{(n-2)!}{(x-2)!(n-x)!}}{(u^2)(u^{x-2})}{(1-u)^{(n-x)}}
\end{align*}

\begin{align*}
= (n)(n-1)(u^2)\sum_{x=1}^n{\frac{(n-2)!}{(x-2)!(n-x)!}}{(u^{x-2})}{(1-u)^{(n-x)}}
\end{align*}

Following the similar process as above, the equation can be written as : 

\begin{align*}
E[X^2 - X] = (n)(n-1)(u^2)
\end{align*}


\begin{align*}
=> E[X^2] - E[X] = (n)(n-1)(u^2)
\end{align*}

Thus, 

\begin{align*}
=> E[X^2] = (n)(n-1)(u^2) + E[X]
\end{align*}

Now, 

Putting in terms of variance : 

\begin{align*}
=> Var(x) = E[X^2] - [E[X]]^2 = (n)(n-1)(u^2) + E[X] - [E[X]]^2
\end{align*}

We have already derived that expected value is nu. Thus, 

\begin{align*}
Var(x) = (n)(n-1)(u^2) + E[X] - [E[X]]^2 = (n)(n-1)(u^2) + (nu) - (nu)^2 = nu(1-u)
\end{align*}
